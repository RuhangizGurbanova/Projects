```python
import pandas as pd
import numpy as np
import requests
import re
from bs4 import BeautifulSoup
```

```python
binaaz=pd.DataFrame()
```

```python
for page_num in range(1,10): 
    url=f'https://bina.az/items/all?page={page_num}'

    page=requests.get(url)

    hmtl=BeautifulSoup(page.content, 'html.parser')

    bina=hmtl.find_all('div', class_='card_params')

    for i in range(len(bina)):
        info=re.sub('<[^>]+>', ' ', str(bina[i])).split('   ')
        if '/gün' in str(info):
            saletype='Günlük'
        elif '/ay' in str(info):
            saletype='Ayliq'
        else:
            saletype='Umumi'

        price= info[1].replace('AZN', '').replace(' ', '')

        location=info[2].strip()

        city=info[4].split(',')[0].strip()

        date=info[4].split(',')[1].strip()

        if info[3].count(' ')==7:
            room_count=info[3].split('  ')[0]
            kvm=info[3].split('  ')[1]
            floor=info[3].split('  ')[2]
        elif info[3].count(' ')==4:
            room_count=info[3].split('  ')[0]
            kvm=info[3].split('  ')[1]
        else:
            kvm=info[3]
        binaaz=binaaz.append({
            'Sale_Type':saletype,
            'Price': price,
            'Location': location,
            'City': city,
            'Rooms_count': room_count,
            'KvM': kvm,
            'Floor': floor,
            'Date': date
        }, ignore_index=True)    
```
