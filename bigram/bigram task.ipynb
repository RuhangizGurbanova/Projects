```python
with open("haber.txt", encoding="utf8") as f:
    lines=f.readlines()
    #print(lines)
    print(type(str(lines)))
    lines.remove('\n')
    print(lines)
    
```

```python
new_lines=[]
for i in lines:
    new_lines.append(i.rstrip())
print(new_lines)
```

```python
bigrams = [b for l in new_lines for b in zip(l.split(" ")[:-1], l.split(" ")[1:])]
print(bigrams)
```
